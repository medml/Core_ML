## Linear Regression (intermediate)
Linear regression pretty much always boils down to "can I fit a line through my data".
In two dimensions, this is easy to visualize.
A little tougher in three dimensions, but once we get to four dimensions and above, it becomes very, very difficult to visualize.

But a lot of the problems we deal with in medicine are *high dimensional*.
In other words: they have a lot of moving parts.

Turns out: linear regression is still a powerful tool no matter how many moving parts you have.
In this notebook we're going to focus on a situation where we have *multiple* independent variables and *a single* dependent variable.
In the language of engineering, we have multiple inputs and a single output.

### Aside: The Experimental Approach
How we would go about designing a set of experiments to *find out* the link between everything?

In a standard approach, we'd find a lab so that we can keep a very carefully controlled environment.
We'd then keep all but two of the variables constant, change one (independent variable), and see if the other one changes (dependent variable).
Because our measurements are never perfect, we would do a power analysis to find out how many times we'd have to do this experiment to be able to reliably say if two factors are/aren't related.

Since there are ~10 factors there, we'd need 10 * 9 = 90 experiments for this.
In each experiment we'd need more than one trial, the exact number depends on how good our measurements are.
Let's be generous and say they're decent, so we only need 5 trials for each experiment.
That leaves us 450 experiments to figure out the structure of the network above.

Or, we could turn to ML.


### Generative Model vs Regression Model (Mathy)

We know the PT ($y$ below) involves certain factors $X_i$ where each factor contributs a certain amount $c_i$.

$y = c_1 \cdot X_1 + c_2 \cdot X_2 + c_3 \cdot X_3 + c_4 \cdot X_4 + c_5 \cdot X_5 + c_6 \cdot X_6 + c_7 \cdot X_7 + c_8 \cdot X_8 + c_9 \cdot X_9 + c_{10} \cdot X_{10} + c_{11} \cdot X_{11} + c_{12} \cdot X_{12} + c_{13} \cdot X_{13}$

But we know that not every factor directly contributes to the PTT, so a lot of the coefficients $c_i$ are actually equal to zero!

$y = c_1 \cdot X_1 + c_2 \cdot X_2 + c_3 \cdot X_3 + c_4 \cdot X_4 + c_5 \cdot X_5 + c_6 \cdot X_6 + c_7 \cdot X_7 + c_8 \cdot X_8 + c_9 \cdot X_9 + c_{10} \cdot X_{10} + c_{11} \cdot X_{11} + c_{12} \cdot X_{12} + c_{13} \cdot X_{13}$

In a much more simplified form, this is what we know the *truth* is:

$y = c_1 \cdot X_1 + c_2 \cdot X_2 + c_3 \cdot X_3 + c_6 \cdot X_6 + c_{10} \cdot X_{10}$

And we know that our measurements are never perfect and that there's *noise*.
Think of noise as 10 of your friends checking their phone to see what the temperature outside in Atlanta is and getting ~5 different numbers -> you know there's only a single number but measurements come with a little 'jitter'.


**So the equation that will be *generating* our data:**
$\text{PTT} = 3 \times \text{XII} + 2.4 \times \text{XI} - 2.0 \times \text{IX} + 3.4 \times \text{VIII} - 2.0 \times \text{I}$

Our goal is to figure out what this equation is knowing only the concentration of factors and the PTT value, and *not* knowing the specific numbers .
That's what the regression will help us find.

## Standard Linear Regression

So, we've got ourselves a dataset from 10 patients.
In each of these 10 patients we measure 14 things: 13 factors and the PTT.
We're asking two questions:
*   Which factors go into the PTT?
*   How much of each of those factors go into the PTT?

One challenge: our data is "smaller" than the number of factors we're looking at.
We have 13+1 variables but only 10 patients.
This can lead to something called *overfitting* where we learn a linear that fits our specific dataset but can be way off for an identical dataset we haven't seen yet.
We're *overdone* the fit of our model to the data we have on hand and not on the *truth* that happens to be generating our model.

Let's see how linear regression does anyway before we start working with some ML concepts.

### Looking at our data
Like any good researcher, we *always* look at the data first.
Let's plot the (mean-subtracted) factor concentrations (y-axis) for each patient (x-axis).

```python
## THIS GENERATES OUR DATA
#Here, we're only including 10 patients, but we have 13 variables that we're trying to study. Would normally be a problem!
X,y = gen_data(pt_len=10)
plot_raw_data(X,y,CoCa)
```


### Setting up the regression

So we've got our data, now we want to try to fit a line to it.

Except this is no ordinary line.
It's a line that lives in 13 dimensions.


```python
# Setup our regression model using LinearRegression() from scikit-learn
reg = sklearn.linear_model.LinearRegression(fit_intercept=True).fit(X.T,y)

plot_models([reg],labels=['Standard Regression'])
```

The linear regression approach looks decent, but it does something that we know can't be true: it thinks pretty much *every* factor contributes to the PTT.
There are some coefficients in the model that are small, but still non-zero (See factor X, VII) and there are some that are included but way off (Factor V, tt, sc).

This can happen when we don't have enough *data* (patients) for the number of moving parts (factors) we're trying to study.

The standard solution here would be to go collect samples from more patients so that we have enough to fit 13 variables *and* overcome the noise that is in our data.
This puts the pressure on data collection which can be slow, expensive, and sometimes impossible.

## REGULARIZED Linear Regression to the rescue!

This is where ML can help.
One way we can overcome the challenge of not having enough data is to use *regularization*.
This involves two things:

1) We still let the computer perform the linear regression 

2) but we also encourage (optimise) it to try to keep certain Factors out of its model.

What this does is it "guides" the computer as it learns the model that accounts for our data.
Usually it's the *data* that does this guiding but, in our case, we don't have enough of it.
Since we also don't know what factors go into the PTT (you forgot, remember?), we can't just ignore certain variables.

Regularization discourages the algorithm from having non-zero coefficients in the final model linking Factors to PT unless it's **really really** useful to include them.
This is a rigorous way that we can tell the algorithm to "look for models that have this specific property that we're pretty sure the real model would have".

```python
lwid = interactive(run_lasso,alpha=(0.01,100.0,0.01))
display(lwid)
#Get our modified regression model from inside the widget
```

Above you'll see a figure that looks familiar but this time we've got a new (green) line.
This line corresponds to the [LASSO regularization]().

In a nutshell, it tells the machine to learn a linear regression model that has as many zeros in it as possible.
How loudly does it tell the machine to do so?
That's the *hyperparameter* called "alpha".
Alpha is a slider that we adjust to discourage non-zero coefficients in our model more and more.
Try it out above.

  1) If you move the slider to the left, you'll find a green line that's much less closer to the blue line (truth) than the orange line (standard linear regression).
  
  2) As you start moving the slider up to ~10.0, you'll see that the green line looks a lot more like the blue line. We have the exact same data, we're just discouraging the algorithm more and more.

  3) If you move it *too far*, let's say around 30.0, we start seeing something unfortunate: we see the green line start shrinking where it used to be right. Take a look at Factor VIII specifically.

## Replicating our findings
Great, so we used LASSO, with an alpha of about ~10.0 to find a pretty good model using only 10 patients worth of data.
Is the model *useful*?
It the model *real*?

Since this is a simulation, we know how far the green model is from the blue model: not that much.
But we want to do a better job of assessing how good the model actually is because this will become important when we *don't* know the truth.

One of the (very correct) warnings that we're taught when analysing data is: you'll *always* find patterns, doesn't mean they're really there.
One way to be sure a pattern is there is to *collect a lot of data* and then look for simple patterns [null-hypothesis testing](), testing them very rigorously.
This assumes that the thing we're studying is just the sum of its parts and most interesting things aren't, by definition.

Another way is to see how well the results from this study do in an entirely new study: we'll do a replication study!
Let's do that now, let's take the model we learned above and see how well we find the same pattern in a new dataset.

**WARNING: this code runs off of the GREEN model you learned *above*. Meaning, if you're trying to see how improved Lasso is over the standard Linear Regression, go back up and set the alpha to ~2.0**
